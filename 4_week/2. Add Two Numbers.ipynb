```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def reverse_list(l1 : ListNode):
    head = l1
    reverse_node = ListNode(val=head.val, next=None)
    node = head.next
    while node != None:
        reverse_node = ListNode(val=node.val, next=reverse_node)
        node = node.next
        if node == None:
            break
        if node.next == None:
            reverse_node = ListNode(val=node.val, next=reverse_node)
            break
    return reverse_node

#linked list to num
def ll2num(node):
    total = 0
    while node != None:
        total = 10 * total + node.val
        node = node.next        
    return total


def print_linked_list(node):
    tmp = node
    while tmp != None:
        print(tmp.val, end= " ")
        tmp = tmp.next
    print()

def print_ll(head):
    while head != None:
        print(head.val, end=" ")
        head = head.next
    print()
        
def num2ll(num : int):
    val = num % 10
    node = ListNode(val=val, next=None)
    num //= 10
    while num != 0:
        val = num % 10
        node = ListNode(val=val, next=node)
        num //= 10
    return node

class Solution:
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
        #reverse_list(l1)
        print_linked_list(l1)
        
        rev1 = reverse_list(l1)
        #print_linked_list(rev)
        num1 = ll2num(rev1)
        print("num1 : ", num1)
        
        rev2 = reverse_list(l2)
        num2 = ll2num(rev2)
        print("num2 : " , num2)
        total = num1 + num2
        node = num2ll(total)
        print_linked_list(node)
        rev_ll = reverse_list(node)
        print_linked_list(rev_ll)
        return rev_ll
```

```python
testcase = [
                [ListNode(2,next=ListNode(4,next=ListNode(3,next=None))),
                ListNode(5,next=ListNode(6,next=ListNode(4,next=None)))]
           ]

solution = Solution()
for i, elem in enumerate(testcase):
    print(i,"th - testcase : ", elem)
    print("input : ", end=" ")
    print_ll(elem[0])
    print_ll(elem[1])
    print("-----")
    res = solution.addTwoNumbers(elem[0],elem[1])
    print("output : ", end=" ")
    print_ll(res)
    
```
