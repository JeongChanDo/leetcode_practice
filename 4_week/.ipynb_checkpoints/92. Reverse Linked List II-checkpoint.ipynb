```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def print_ll(head):
    while head != None:
        print(head.val, end =" ")
        head = head.next
    print()
    

    
def reverse_ll(head):
    
    node = head.next
    rev = ListNode(val=head.val, next=None)
    while node != None:
        rev = ListNode(val=node.val, next=rev)
        node = node.next
    return rev

class Solution:
    def reverseBetween(self, head: ListNode, m: int, n: int) -> ListNode:
        if m == n:
            return head
        
        cursor = head
        
        before_m = None
        mth_node = None
        after_m = None
        
        before_n = None
        nth_node = None
        after_n = None
        
        idx = 1
        while cursor != None:
            
            if idx + 1 == m:
                before_m = cursor
                mth_node = cursor.next
                after_m = cursor.next.next
            elif idx == 1 and m == 1:
                mth_node = cursor
                after_m = cursor.next
            
            if idx + 1 == n:
                before_n = cursor
                nth_node = cursor.next
                after_n = cursor.next.next
                break

            cursor = cursor.next
            idx +=1
        print("after_n : ", end="")
        print_ll(after_n)


        new_head = mth_node
        new_cursor = mth_node
        while new_cursor != None:
            if new_cursor == nth_node:
                new_cursor.next = None
                break
            new_cursor = new_cursor.next
        rev = reverse_ll(new_head)
        print("rev : ", end="")
        print_ll(rev)


        if m != 1:
            before_m.next = rev
        else:
            head = rev
        rev_cursor = rev
        while rev_cursor != None:
            if rev_cursor.next == None:
                rev_cursor.next = after_n
                break
            rev_cursor = rev_cursor.next
        print("starts with : ", end="")
        print_ll(head)
        print("-----")
        return head
```

```python
testcase = [
            [
                ListNode(1,next=ListNode(2,next=ListNode(3,next=ListNode(4,next=ListNode(5,next=None))))),
                2, 4]
           ]

solution = Solution()
for i, elem in enumerate(testcase):
    print(i,"th - testcase : ", elem)
    print("input : ", end=" ")
    print_ll(elem[0])
    print("-----")
    res = solution.reverseBetween(elem[0], elem[1], elem[2])
    print("output : ", end=" ")
    print_ll(res)
    
```
