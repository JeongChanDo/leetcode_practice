```python
class Solution:
    def simplifyPath(self, path: str) -> str:
        """
        5 steps for simplify path
        
        1. delete multiple slash
        2. get all word
            - delete all last slash
            - add slash to end of the path
            - get words between slashes
            - delete last slash again
        3. delete some dot '.' that means corrent directory
        4. get current directory with list using append & pop
        5. complete simpliest path
        """
        print("###",path,"###")
        #delete multiple slash
        tmp_lst = []
        for i in range(len(path)):
            if i + 1 < len(path):
                if path[i] == '/' and path[i+1] == '/':
                    continue
            tmp_lst.append(path[i])
        slash_str = ''.join(tmp_lst)
                
        #delete slash at the end of the path
        if slash_str[-1] == '/':
            slash_str = slash_str[:-1]
        print("slash str before get word : ", slash_str)
        
        #add last slash to get all words between slashes
        slash_str = slash_str + "/"
        
        
        
        #get words betweens slash
        start_slash_idx = 0
        end_slash_idx = 0
        words = []
        for i in range(len(slash_str)):
            if slash_str[i] == '/' and i == 0:
                start_slash_idx = i
            elif slash_str[i] == '/':
                end_slash_idx = i
                word = slash_str[start_slash_idx + 1:end_slash_idx]
                words.append(word)
                start_slash_idx = i
        print("words : ", words)
        
        
        #delete slash at the end of the path again
        if slash_str[-1] == '/':
            slash_str = slash_str[:-1]
        print("slash str after get words : ", slash_str)
        
        #delete '.' that means current directory
        while '.' in words:
            words.remove('.')
        
        #get current directory
        directories = []
        for i in range(len(words)):
            if words[i] == "..":
                if not directories:
                    pass
                else:
                    directories.pop()
            else:
                directories.append(words[i])
        print("current directory : ", directories)
        
        
        #get simple path
        simple_path = "/"
        for i in range(len(directories)):
            simple_path = simple_path + directories[i]
            if i + 1 != len(directories):
                simple_path = simple_path + "/"
        print("simple_path : ", simple_path)
        print()
        return simple_path
```

```python
testcase = ["/home/",
        "/../",
        "/home//foo/",
        "/a/./b/../../c/"]

solution = Solution()
for i, string in enumerate(testcase):
    print(i,"th - testcase : ", string)
    print("result : ", solution.simplifyPath(string))
    print()
    
```
