```python
class MyQueue:
    
    def __init__(self):
        self.queue = []
    
    
    def push(self, x: int) -> None:
        self.queue.append(x)
    
    
    def peek(self) -> int:
        return self.queue[0]
    
    
    def pop(self):
        peek = self.peek()
        self.queue.remove(peek)
        return peek
    
    
    def size(self) -> int:
        return len(self.queue)   
    
    
    def is_empty(self) -> bool:
        if self.queue:
            return False
        return True

    
class MyStack:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.queue1 = MyQueue()
        self.queue2 = MyQueue()

        
    def push(self, x: int) -> None:
        """
        Push element x onto stack.
        """
        while not self.queue1.is_empty():
            self.queue2.push(self.queue1.pop())
        self.queue1.push(x)
        while not self.queue2.is_empty():
            self.queue1.push(self.queue2.pop())

    def pop(self) -> int:
        """
        Removes the element on top of the stack and returns that element.
        """
        return self.queue1.pop()

    def top(self) -> int:
        """
        Get the top element.
        """
        return self.queue1.peek()

    def empty(self) -> bool:
        """
        Returns whether the stack is empty.
        """
        return self.queue1.is_empty()
        
```
